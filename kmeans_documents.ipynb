```
import glob
onlyfiles=glob.glob("/content/drive/My Drive/Colab Notebooks/test/*.txt")
onlyfiles
```

```
documents = []
```

```
for file in onlyfiles:
  with open(file, 'r') as myfile:
    data = myfile.read()
  documents.append(data)
documents
```

```
nltk.download('punkt')
```

```
combined = ''
for i in documents:
  combined+=i
text1 = combined.split(' ')
fdist1 = nltk.FreqDist(text1)
print(fdist1.most_common(50))
```

```
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
```

```
stopwords
```

```
tokens = [word.lower() for doc in documents for sent in nltk.sent_tokenize(doc) for word in nltk.word_tokenize(sent) if len(word)>3 and word not in stopwords]
set(tokens)
```

```
vectorizer = CountVectorizer(vocabulary = set(tokens))

X = vectorizer.fit_transform(documents)

print(X.toarray())
```

```
true_k = 2

model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)

model.fit(X)
```

```
print("Top terms per cluster:")

order_centroids = model.cluster_centers_.argsort()[:, ::-1]

terms = vectorizer.get_feature_names()

for i in range(true_k):

    print("Cluster %d:" % i),

    for ind in order_centroids[i, :10]:

        print(' %s' % terms[ind]),

    print
```

```
t=model.labels_
t
```

```
for n in range(true_k):
  print("Cluster",n,':\n')
  for j,i in enumerate(t):
    if i==n:
      print(f'Doc{j}\n') 
```
